Create a function that calculates the area of a circle and returns it.

`(Area = 3.14 √ó radius √ó radius)`

**Output:**
```
Circle area: 78.5
Larger circle area: 314.0
Combined area: 78.26

````python
# Solution:
import math as m
def AreaofCir(R1,R2):
  CA=m.pi*(R1*R1)
  LC=m.pi*(R2*R2)
  ComA=CA+LC
  print(f"Circle 1 area {CA}")
  print(f"Circle 2 area {LC}")
  print(f"Combined area {ComA}")

AreaofCir(5,6)



````
